```python
import numpy as np
from scipy.stats import shapiro
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
dados_amostra=np.random.exponential(size=1000)
stat, p_valor=shapiro(dados_amostra)
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.histplot(dados_amostra,kde=True,color='skyblue')
plt.title('Histograma')
plt.subplot(1,2,2)
res=plt.probplot(dados_amostra,plot=plt)
plt.tight_layout()
plt.show()
alpha=0.5
if p_valor > alpha :
    print('Os dados estao normalizados')
else:
    print('Os dados nao estao normalizados')
```
