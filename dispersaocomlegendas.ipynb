```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.rand(50)
y=2*x + 1 +0.1*np.random.randn(50)
plt.scatter(x,y,label='Dados de Exemplo', color='blue', marker='o')
plt.title('Dispersao com Legendas')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend(loc='upper left')
plt.show()
```
